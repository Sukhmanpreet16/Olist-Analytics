```python
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    database="olist",
    user="postgres",
    password="1234"
)
cur = conn.cursor()

base = r"C:/Users/toors/Olist/cleaned"

tables = {
    "orders": f"{base}/orders.csv",
    "customers": f"{base}/customers.csv",
    "sellers": f"{base}/sellers.csv",
    "products": f"{base}/products.csv",
    "order_items": f"{base}/order_items.csv",
    "payments": f"{base}/payments.csv",
    "reviews": f"{base}/reviews.csv",
    "geolocation": f"{base}/geolocation.csv",
    "category_name_translation": f"{base}/category_name_translation.csv"
}

for table, path in tables.items():
    with open(path, "r", encoding="utf-8") as f:
        cur.copy_expert(
            f"COPY {table} FROM STDIN WITH CSV HEADER",
            f
        )

conn.commit()
cur.close()
conn.close()
```
